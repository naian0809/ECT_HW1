**第11題請將剩下的欄位整理成正確的形式**

```python
#想說後面每個資料都是佔一格，於是將所有非空值的資料都向左對齊，為避免影響後排已經排好的，範圍設在第2欄至第19行
import numpy as np
data = df.copy(deep=True)
cols_to_process = data.iloc[:,1:19]
cols_to_process = cols_to_process.apply(lambda x: sorted(x, key=pd.isna), 1).to_list()

data.iloc[:, 1:19] = cols_to_process
```

```python

data['Revenue1'] = data['type.1']
data['Planned revenue'] = data['Retailer']
data['Product cost'] = data['country']
data['Quantity1'] = data['Revenue']
data['Unit cost'] = data['Planned']
data['Unit price'] = data['revenue']
data['Gross profit'] = data['Product.3']
data['Unit sale price'] = data['cost']
data = data.drop(['type.1','Retailer','country','Revenue','Planned','revenue','Product.3','cost','Quantity','Unit','cost.1','Unit.1','price','Gross','profit','Unit.2','sale','price.1'], axis=1)
data.rename(columns={'Revenue1': 'Revenue', 'Quantity1': 'Quantity'}, inplace=True)
```
